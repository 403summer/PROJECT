## 2. 1번 추천 곡 오디오 피처를 바탕으로 곡 추천

```python
# 모델링으로부터 시드 값 추출
seed1 = track_df[track_df['track_id'] == rec_1]['artist_id']
seed2 = track_df[track_df['track_id'] == rec_1]['artist_genres']
seed2 = seed2.str.replace("'",'')
seed2 = seed2.str.replace("[",'')
seed2 = seed2.str.replace("]",'')
seed2 = seed2.str.split(',')
seed2 = list(seed2)
seed3 = rec_1
```

```python
# 시드 - 오디오피처 추출
danceability = audio_df[audio_df['id'] == rec_1]['danceability']
energy = audio_df[audio_df['id'] == rec_1]['energy']
speechiness = audio_df[audio_df['id'] == rec_1]['speechiness']
acousticness = audio_df[audio_df['id'] == rec_1]['acousticness']
instrumentalness = audio_df[audio_df['id'] == rec_1]['instrumentalness']
valence = audio_df[audio_df['id'] == rec_1]['valence']
tempo = audio_df[audio_df['id'] == rec_1]['tempo']
```

```python
#스포티파이 api에 시드값을 넣어 추천
rec = sp.recommendations(seed_artists=[seed1.iloc[0]], seed_genres=[seed2[0][0]], seed_tracks=[seed3], min_popularity=50, max_energy = energy, limit=1)
for track in rec['tracks']:
    print(track['artists'][0]['name'], track['name'])
```
